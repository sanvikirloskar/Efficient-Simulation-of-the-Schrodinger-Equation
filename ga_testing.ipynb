## Genetic Algorithm simulation of the Shaken Lattice
First, import all libraries

```python
import numpy as np
from blochstate1d_NEW import OLConstants, GroundBlochState
from ga_tools import fitness_finder
from genetic_algorithm import eugenics, one_point_crossover, two_point_crossover, mutate, creep, new_gene_maker, full_genetic_algorithm
```

Initialise constants

```python
# Define parameters for the genetic algorithm
constants = OLConstants()
N_individuals = OLConstants.N_individuals  # Number of individuals in the population
N_elements = OLConstants.N_elements      # Number of elements in the control function (e.g., number
sigma = 100 #standard deviation for the normal distribution to generate the coefficients from
mom_pop_des = constants.desired_mom_pop #desired momentum population from the constants file
mom_pop_des = np.array(mom_pop_des) #convert to numpy array for easier calculations
number_dead = 4 #number of individuals to remove each generation

#Initialise starting coeffs for guess shaking function
A_coeffs = np.random.normal(0, sigma, (N_individuals, N_elements)) #generating the A coefficients from a normal distribution
B_coeffs = np.random.normal(0, sigma, (N_individuals, N_elements)) #generating the B coefficients from a normal distribution
omegas = np.random.uniform(0, 0.6, (N_individuals, N_elements)) #generating the frequencies from a uniform distribution between 0 and 0.6
```

```python
#run the first propagation of the code to get fitnesses
fitnesses, mom_pop_first_iter = fitness_finder(A_coeffs, B_coeffs, omegas, mom_pop_des)
```

Now, we have 40 individuals, each with different fitness functions. The next step is to sort them 



```python
best_A, best_B, best_omegas, remaining_A, remaining_B, remaining_omegas, unfit_parents = eugenics(fitnesses, A_coeffs, B_coeffs, omegas, number_dead)
```

Now we have the best A and B coeffs as well as the remaining ones (here, the remaining ones do not include the four worst that we eliminated) so we now have the tools to make 8 new individuals. We should now have N_individuals - 2 - 4 individuals left (after iteration 1, this should be 40-2-4 = 34). Pair them up and allocate either one point or two point crossovers (we do need to consider mutation and creep but i dont know what to do if we end up with an odd number of individuals so we cant have parents)

```python
#print(remaining_A, remaining_B, remaining_omegas)

#randomly pairing them up
n_for_pairing = remaining_A.shape[0]
indices = np.random.permutation(n_for_pairing)
shuffled_A = remaining_A[indices]
shuffled_B = remaining_B[indices]
shuffled_omegas = remaining_omegas[indices]

for i in range(0, n_for_pairing, 2):
    parent1_A = shuffled_A[i]
    parent2_A = shuffled_A[i+1]
    parent1_B = shuffled_B[i]
    parent2_B = shuffled_B[i+1]
    parent1_omegas = shuffled_omegas[i]
    parent2_omegas = shuffled_omegas[i+1]
    
childA, childB, child_omegas = two_point_crossover(parent1_A, parent2_A, parent1_B, parent2_B, parent1_omegas, parent2_omegas)

    #both one and two point crossoves work. 
```

```python
#checking mutation and creep work
mutated_childA, mutated_childB, mutated_child_omegas, unfit_parents = mutate(childA, childB, child_omegas)
creeped_childA, creeped_childB, creeped_child_omegas, unfit_parents = creep(mutated_childA, mutated_childB, mutated_child_omegas, unfit_parents)
```

```python
# testing create new gene
newgen_A, newgen_B, newgen_omegas = new_gene_maker(remaining_A, remaining_B, remaining_omegas, best_A, best_B, best_omegas, unfit_parents, number_dead)
# cool so the new gene maker also works. now in a separate.py file i will combine all this
```
