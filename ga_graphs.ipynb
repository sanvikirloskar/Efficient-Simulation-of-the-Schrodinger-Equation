```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from errors_momentum_space import momentum_space_population
from blochstate1d_NEW import GroundBlochState, OLConstants
```

```python
#formatting the data so it suits our graph. lets remove all generation columns except for 8 kids(control) and combine the dataframes
mr01 = pd.read_csv("GA_controls.csv", header= None)
mr02 = pd.read_csv("GA_4_dead.csv", header= None)
mr03 = pd.read_csv("GA_6_dead.csv", header= None)
mr04 = pd.read_csv("GA_10_dead.csv", header= None)
mr05 = pd.read_csv("GA_12_dead.csv", header = None)
mr06 = pd.read_csv("GA_4_dead_2.csv", header = None)
mr07 = pd.read_csv("GA_4_dead_3.csv", header = None)
mr08 = pd.read_csv("GA_4_dead_4.csv", header = None)
```

```python


# put kids_8 first, then the others
mr_all = pd.concat([mr01, mr02, mr03, mr04, mr05,mr06, mr07, mr08], axis=1)

mr_all.head()
```

```python
# give each column a name matching the number of kids it represents
mr_all.columns = ['8 replaced(control)', '4 replaced','6 replaced', '10 replaced', '12 replaced','4 replaced(2)', '4 replaced(3)', '4 replaced(4)']
colours = ['royalblue','crimson','forestgreen','darkorange','mediumpurple','crimson','crimson','crimson']
# simple line plot with generation on the x‑axis
mr_all.plot(color=colours)
plt.xlabel('Generation', fontsize = 14)
plt.ylabel('Fitnesses', fontsize = 14)
plt.title('GA learning when the population replaced varies', fontsize = 16)
plt.savefig('ga_popdead_final.png', dpi=300)
```

```python
#mutation rate graphs
mutr1 = pd.read_csv("GA_mutation_rate_0.2.csv", header= None)
mutr2 = pd.read_csv("GA_mutation_rate_0.3.csv", header= None)
mutr3 = pd.read_csv("GA_mutation_rate_0.2_2.csv", header= None)
mutr4 = pd.read_csv("GA_mutation_rate_0.2_3.csv", header= None)
mutr5 = pd.read_csv("GA_mutation_rate_0.2_4.csv", header= None)
mutr6 = pd.read_csv("GA_mutation_rate_0.2_5.csv", header= None)
mutrc = pd.read_csv("GA_controls.csv", header= None)
```

```python
#put everything in graphable format
mutr_all = pd.concat([mutrc, mutr1, mutr2, mutr3, mutr4, mutr5, mutr6], axis=1)
mutr_all.columns = ['control','mutation rate 0.2','mutation rate 0.3','mutation rate 0.2(2)', 'mutation rate 0.2(3)', 'mutation rate 0.2(4)', 'mutation rate 0.2(5)']
mutr_colours = ['royalblue', 'crimson', 'forestgreen', 'crimson', 'crimson', 'crimson', 'crimson']
```

```python
#plot
mutr_all.plot(color=mutr_colours)
plt.xlabel('Generation', fontsize = 14)
plt.ylabel('Fitnesses', fontsize = 14)
plt.title('GA learning when the mutation rate varies', fontsize = 16)
plt.legend()
plt.savefig('ga_mutrate_final.png', dpi=300)
```
